### Elementos suplementarios

Para especificar individuos y variables suplementarios, la función MCA se puede usar de la siguiente manera

```r
require(FactoMineR)
require(factoextra)
data(poison)
res.mca2 <- MCA(poison, ind.sup = 53:55, 
               quanti.sup = 1:2, quali.sup = 3:4,  graph=FALSE) 
```

```r
# Supplementary qualitative variable categories
res.mca2$quali.sup
# Supplementary quantitative variables
res.mca2$quanti
# Supplementary individuals
res.mca2$ind.sup
```

#### Biplot con variables suplementarias

```r
# Biplot of individuals and variable categories
fviz_mca_biplot(res.mca2, repel = TRUE,
                ggtheme = theme_minimal())
```

* Las personas activas están en azul
* Los individuos suplementarios están en azul oscuro
* Las categorías de variables activas están en rojo
* Las categorías de variables suplementarias están en verde oscuro

#### Las variables cuantitativas suplementarias

```r
fviz_mca_var(res.mca2, choice = "quanti.sup",
             ggtheme = theme_minimal())
```

 Visualizando los individuos suplementarios se tiene:

```r
 fviz_mca_ind(res.mca2, 
             label = "ind.sup", #Show the label of ind.sup only
             ggtheme = theme_minimal())
```

#### GRÁFICA PARA UNA CANTIDAD DE VARIABLES E INDIVIDUOS ALTA (RESULTADOS DE FILTRADO)

```r
# Visualize variable categories with cos2 >= 0.4
# Top 10 active variables with the highest cos2
fviz_mca_var(res.mca2, select.var= list(cos2 = 10))
```

```r
fviz_mca_biplot(res.mca2, select.ind = list(contrib = 5), 
                select.var = list(contrib = 5),
                ggtheme = theme_minimal()) # top 5 contributing individuals and variable categories
```
